```python
import cv2
import matplotlib.pyplot as plt
```

```python
# load in already built model with its configuration file
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

```python
# load model into openCV
model = cv2.dnn_DetectionModel(frozen_model, config_file)
```

```python
# set model configurations
model.setInputSize(320, 320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5, 127.5, 127.5))
model.setInputSwapRB(True) # this automatically changes image from BGR to RGB
```

```python
#Â import labels into list
classLabels = []
with open('Labels.txt', 'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

# read in image


```python
image = cv2.imread('man_bmw.jpg')
```

```python
# convert image from BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image) # bgr image
```

```python
# make prediction on single image
ClassIndex, confidece, bbox = model.detect(image, confThreshold=0.5)
```

```python
# print out detect details
print(ClassIndex)
print(confidece)
print(bbox)
```

```python
# show section of image with detection
font_scale = 3
font = cv2.FONT_HERSHEY_SIMPLEX

for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
    cv2.rectangle(image, boxes, (255, 0, 0), 2)
    cv2.putText(image, classLabels[ClassInd-1], (boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(0, 255, 0), thickness=3)
```

```python
plt.imshow(image)
```

```python
# video caption
# cap = cv2.VideoCapture('Pexels Videos 1721294.mp4')
cap = cv2.VideoCapture(0)

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot Open Video")
    
font_scale = 2
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()
    
    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)
    print(ClassIndex)
    
    if len(ClassIndex) != 0:
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if ClassInd <=80:
                cv2.rectangle(frame, boxes, (255, 0, 0), 2)
                cv2.putText(frame, classLabels[ClassInd-1], (boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(0, 255, 0), thickness=3)
    cv2.imshow('Detecting...', frame)
    
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
                

# while cap.isOpened():
#     ret_val, frame = cap.read()
#     if ret_val:
#         # make detection
#         ClassIndex2, confidence2, bbox2 = model.detect(frame, confThreshold=0.55)
#         for ClassInd, conf, boxes in zip(ClassIndex2.flatten(), confidence2.flatten(), bbox):
#             if ClassInd <= 80:
#                 cv2.rectangle(frame, boxes, (255, 0, 0), 2)
#                 cv2.putText(frame, classLabels[ClassInd-1], (boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(0, 255, 0), thickness=3)        
#         cv2.imshow('Detecting...',frame)
#         if cv2.waitKey(1) & 0xFF == ord('q'):
#             break
#         else:
#             break
        
#     else:
#         break

cap.release()
cv2.destroyAllWindows()
        
```
